```python
import csv, sqlite3

with open("./GraduateEmploymentSurveyNTUNUSSITSMUSUSSSUTD.csv", "r") as f:
    csv_reader = csv.reader(f)
    header = next(csv_reader)
    ph = ",".join("?" for _ in header)
    data = []
    for line in csv_reader:
        data.append(line)
query = """
CREATE TABLE IF NOT EXISTS GraduateEmploymentSurvey(
	"year"	INTEGER,
	"university"	TEXT,
	"school"	TEXT,
	"degree"	TEXT,
	"employment_rate_overall"	REAL,
	"employment_rate_ft_perm"	REAL,
	"basic_monthly_mean"	INTEGER,
	"basic_monthly_median"	INTEGER,
	"gross_monthly_mean"	INTEGER,
	"gross_monthly_median"	INTEGER,
	"gross_mthly_25_percentile"	INTEGER,
	"gross_mthly_75_percentile"	INTEGER,
    PRIMARY KEY("year","university","school","degree")
);
"""
conn = sqlite3.connect("./survey.db")
conn.execute(query)
conn.commit()
query2 = f"""
INSERT INTO GraduateEmploymentSurvey
{tuple(header)}
VALUES
({ph})
"""
for line in data:
    conn.execute(query2, tuple(line))
conn.commit()
conn.close()
```
